Now what I have done is created a dataframe that shows, playlist name, artist name, track name, album name, and the top artist count. We can see the number on the far right shows us the amount of times those top artist for each playlist apears in all of the listed playlist.

```python
count = 0
for x in data_df['song'][4]:
    artist = x['track']['artistName']
    if (artist == 'Drake'):
        count += 1
print(count)
```

The line above shows the algorithm for the table above to figure out how many times a certain artist appears on one of the 17 playlist.

```python
import numpy as np
```

Im importing numpy package

```python
def artist_counter(playlist, artist):
    count = 0 
    for x in playlist:
        if (x['track']['artistName'] == artist):
            count +=1
    return(count)
```

```python
data_df['top_artist_count'] = np.nan
```

```python
for i in range(0, len(data_df.index)):
    data_df['top_artist_count'][i] = artist_counter(data_df['song'][i], data_df['artistName'][i])
    data_df
```

So the three lines above is me getting ready to setup a visual to display the top artist in each playlist.  The first one is another algorithm that will help us calculate the top artist in all the playlist all at once.

```python
from matplotlib import pyplot as plt
plt.style.use('dark_background')
```

```python
fig, ax0 = plt.subplots(figsize=(20,15))
ax0.bar(data_df[:17]['artistName'], data_df['top_artist_count'])
plt.xticks(rotation =60, color = 'white', fontsize = 20)
plt.xlabel('Artist', fontsize = 20)
plt.yticks( color = 'white', fontsize = 30)
plt.ylabel('Number of artist who show up in playlist', fontsize = 20)
plt.title("Top Artist Count", fontsize = "30", color = "white")
plt.show()
```

Finally we have out visual bar grapch and it shows up all of my top artist for each playlist.
