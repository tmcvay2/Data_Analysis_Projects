```python
import pandas as pd
import warnings
```

```python

oly_uniq = pd.read_csv('Olympics_Postponed_Twitter_Unique.csv')
oly_uniq.keys()
```

```python
oly_uniq['Language'].head(20)
```

```python
oly_uniq["Full Text"][12]
```

```python
en_only = oly_uniq.loc[oly_uniq['Language'] == 'en']
en_only
```

```python
en_only['Full Text'].head()
```

```python
en_only['text_lower'] = [x.lower() for x in en_only['Full Text']]
en_only['text_lower'].head(20)
```

```python
warnings.filterwarnings('ignore')
```

```python
en_only['Full Text'][1]
```

```python
import nltk
from nltk import tokenize as tk
from nltk import FreqDist
from nltk.corpus import stopwords
import matplotlib
import string
nltk.download('punkt')
nltk.download('stopwords')
```

```python
en_only['tokens'] = [tk.word_tokenize(x) for x in en_only['text_lower']]
en_only['tokens'].head(5)
```

```python
en_only['tokens'][2]
```

```python
len(stopwords.words('english'))
```

```python
mystopwords = stopwords.words('english')
mystopwords.append("n't")
mystopwords.append("'s")
len(mystopwords)
```

```python
filtered = []
for i in en_only["tokens"]:
    filtered.append([word for word in i if word not in mystopwords])
    
en_only['filtered'] = filtered
```

```python
en_only['filtered'][2]
```

```python
string.punctuation
```

```python
no_punct = []
for i in en_only['filtered']:
    no_punct.append([word for word in i if word not in string.punctuation])

en_only['no_punct'] = no_punct
en_only['no_punct'][2]
```

```python
freq1 = [FreqDist(x) for x in en_only['no_punct']]
en_only['freqs1'] = freq1
en_only['freqs1']
```

```python
from collections import Counter
```

```python
A = Counter(en_only['freqs1'][0])
B = Counter(en_only['freqs1'][1])
A + B
```

```python
en_only['f1_counts'] = [Counter(x) for x in en_only['freqs1']]
en_only['f1_counts'][0]
```

```python
wordcounts = Counter()
for i in en_only['f1_counts']:
    wordcounts += i
```

```python
len(wordcounts)
```

```python
type(wordcounts)
```

```python
dict(wordcounts)
```

```python
counts = pd.DataFrame.from_dict(dict(wordcounts), orient ='index')
counts
```

```python
counts = counts.rename(columns = {0: 'counts'})
counts
```

```python
count_sort = counts.sort_values('counts', ascending = False)[:40]
count_sort
```

```python
from matplotlib import pyplot as plt
```

```python
fig, ax0 = plt.subplots(figsize=(15,10))
ax0.bar(count_sort.index, count_sort['counts'])
plt.xticks(rotation=90, fontsize = 'small', color = 'white')
plt.yticks(fontsize = 'small', color = 'white')
plt.show()
```

```python
stemmer = nltk.stem.SnowballStemmer("english")
stems = []
for i in en_only['no_punct']:
    stems.append([stemmer.stem(word) for word in i])
    
en_only['stemmed'] = stems
```

```python
freq2 = [FreqDist(x) for x in en_only['stemmed']]
en_only['freqs2'] = freq2
en_only['freqs2']
```

```python
en_only['f2_counts'] = [Counter(x) for x in en_only['freqs2']]
en_only['f2_counts'][0]
```

```python
wordcounts = Counter()
for i in en_only['f2_counts']:
    wordcounts += i
```

```python
counts2 = pd.DataFrame.from_dict(dict(wordcounts), orient='index')
counts2
```

```python
counts2 = counts2.rename(columns = {0: 'counts'})
counts2
```

```python
counts_sort2 = counts2.sort_values('counts', ascending = False)[:20]
counts_sort2
```

```python
fiq, ax1 = plt.subplots(figsize=(10,10))
ax1.bar(counts_sort2.index, counts_sort2['counts'])
plt.xticks(rotation=90, fontsize = 'medium', color = 'white')
plt.yticks(fontsize = 'medium', color = 'white')
plt.show()
```
